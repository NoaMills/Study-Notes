First we install matplotlib, pyplot, and numpy.

```python
import matplotlib
print("Matplotlib version: ", matplotlib.__version__)
import matplotlib.pyplot as plt
import numpy as np
```

We can create a simple plot using the plt.plot(x,y) function. This takes as arguments arrays x and y, and creates a plot connecting the given points with a line.

```python
xpoints=np.array([0,6,8])
ypoints=np.array([1,4,2])
plt.plot(xpoints, ypoints)
plt.show()
```

If no x axis values are given, we assume the x axis values are \[0,1,2,3...\] fitting the size of the y axis values.

```python
ypoints = np.array([5,2,3,5,8,2,1])
plt.plot(ypoints)
plt.show()
```

If we wish to plot points without connecting lines, we can use the "shortcut string notation" parameter 'o' to specify that we wish to plot rings.

```python
plt.plot(ypoints, 'o')
plt.show()
```

We can also create a plot with both lines and dot markers by using the keyword "markers" as follows. More marker options are listed here: https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(ypoints, marker='o')
plt.show()
plt.plot(ypoints, marker='*')
plt.show()
```

We can specify the marker, line type, and color all in one shortcut string notation parameter. These string parameters use characters to specify marker|line|color in that order. In the example below, ':' signifies a dotted line and 'r' signifies red. More line types and colors are listed here: https://matplotlib.org/1.5.3/api/pyplot_api.html#matplotlib.pyplot.plot If no line specification is included, then the markers are not connected by a line.

```python
plt.plot(ypoints, 'o:r')
plt.show()
plt.plot(ypoints, "xk")
plt.show()
```

We can modify the marker size using the "ms" parameter as follows.

```python
plt.plot(ypoints, "ok", ms=20)
plt.show()
plt.plot(ypoints, "ok", ms=2)
plt.show()
```

We can use markeredgecolor (or mec for short) to specify the color of the edge of the markers.

```python
plt.plot(ypoints, "o:y", mec='b', ms=16)
plt.show()
```

Likewise, we can use markerfacecolor (or mfc) to specify the color of the inside of the markers.

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r') #default color is blue
plt.show()
```

We can set both mec and mfc to make the entire marker a specific color without affecting the line color

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r', mec = 'r')
plt.show()
```

We can also specify a color using hex color values or one of the 140 color names listed here: https://www.w3schools.com/colors/colors_names.asp

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = 'HoneyDew')
plt.show()
```

We can use the keyword argument linestyle (or ls for short) to specify if we want the line to be dotted, dashed, or solid.

```python
plt.plot(ypoints, linestyle="dotted", marker='o', ms=10) #or ls=':'
plt.show()
plt.plot(ypoints, linestyle="dashed", marker='o', ms=10) #ls="--"
plt.show()
plt.plot(ypoints, linestyle="solid", marker='o', ms=10) #ls="-"
plt.show()
plt.plot(ypoints, linestyle="dashdot", marker='o', ms=10) #ls = "-."
plt.show()
plt.plot(ypoints, linestyle="None", marker='o', ms=10) #ls=""
plt.show()
```

We can use the keyword argument "color" (or "c") to specify the color of the line.

```python
plt.plot(ypoints, c="r")
plt.show()
plt.plot(ypoints, c="#95EFA1")
plt.show()
```

We can use the keyword argument "linewidth", (or lw) to specify the width of the line.

```python
plt.plot(ypoints, c="r", lw=1)
plt.show()
plt.plot(ypoints, c="r", lw=10)
plt.show()
```

We can draw two lines on the same plot by calling plt.plot() twice.

```python
firstline = np.array([5,2,1,4,7])
secondline = np.array([1,2,5,2,3])
plt.plot(firstline, c="Aqua")
plt.plot(secondline, c="DarkOrange")
plt.show()
```

Alternatively, we can create a plot with multiple lines by passing the pairs of x and y arrays to the same plt.plot function. <b> Question:</b> Can we still specify colors for each line with this method?

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

We can use the xlabel() and ylabel() functions to label our axes. We can use the title() function to add a title to the plot.

```python
plt.plot(ypoints, 'o:r')
plt.xlabel("I am an x axis!")
plt.ylabel("I am a y axis!")
plt.title("Title mcTitleFace")
plt.show()
```

We can modify the text properties of the title and axis labels directly, or by creating a dictionary of font parameters and values, and setting the "fontdict" property to this dictionary.

```python
font1={"color":"black", "size":"20"}
plt.plot(ypoints, 'o:r')
plt.xlabel("I am an x axis!", size = 20, color = 'blue')
plt.ylabel("I am a y axis!", fontdict=font1)
plt.title("Title mcTitleFace", font1)
plt.show()
```

We can set the position of the title to the left, right, or center using the "loc" argument.

```python
plt.plot(ypoints, 'o:r')
plt.xlabel("I am an x axis!")
plt.ylabel("I am a y axis!")
plt.title("Title mcTitleFace", loc="left")
plt.show()
```

We can add a grid to our plot with the plt.grid() function. By default, we show gridlines for both the x and y axis. If we wish to show gridlines for just one axis, we can specify the axis we would like to show as follows.

```python
plt.plot(ypoints, 'o:r')
plt.xlabel("I am an x axis!")
plt.ylabel("I am a y axis!")
plt.title("Title mcTitleFace", loc="left")
plt.grid()
plt.show()

plt.plot(ypoints, 'o:b')
plt.xlabel("I am an x axis!")
plt.ylabel("I am a y axis!")
plt.title("Title mcTitleFace", loc="left")
plt.grid(axis="x")
plt.show()
```

We can also stylize our gridlines as we would our plotted lines.

```python
plt.plot(ypoints, 'o:r', lw=3)
plt.xlabel("I am an x axis!")
plt.ylabel("I am a y axis!")
plt.title("Title mcTitleFace", loc="left")
plt.grid(ls=':', c="b", lw=1.5)
plt.show()
```

We can use the subplots function to plot multiple figures in one image. The subplot function takes three arguments: rows, columns, and index of current plot.

```python
y1=np.array([4,3,5,6,2,40])
y2=np.array([8,6,6,5,4,2,1,1,1,3])

plt.subplot(1,2,1)
plt.plot(y1, c="r")
plt.grid()
plt.subplot(1,2,2)
plt.plot(y2, c="b")
plt.grid()
plt.show()
```

We can title each figure individually with the title() function, and the entire image with the suptitle() function

```python
y1=np.array([4,3,5,6,2,40])
y2=np.array([8,6,6,5,4,2,1,1,1,3])

plt.subplot(1,2,1)
plt.plot(y1, c="r")
plt.grid()
plt.title("Red plot")
plt.subplot(1,2,2)
plt.plot(y2, c="b")
plt.grid()
plt.title("Blue plot")
plt.suptitle("Red plot and blue plot")
plt.show()
```

We can create a scatter plot with the plt.scatter() function.

```python
x1=np.random.randint(0,101,50)
y1=np.random.randint(0,51,50)
x2=np.random.randint(0,101,50)
y2=np.random.randint(25,61,50)
plt.scatter(x1,y1, color = "LightCoral")
plt.scatter(x2,y2, color="LightSeaGreen")
plt.show()
```

We can specify colors for each point using the "c" argument but NOT the "color" argument.

```python
colorArray=np.array(["OrangeRed", "LightBlue", "LightCoral", "LightSeaGreen", "MediumBlue", "OliveDrab"])
xpoints=np.array([0,1,2,3,4,5])
ypoints =np.array([3,4,3,6,7,7])
plt.scatter(xpoints, ypoints, c=colorArray)
plt.show()
```

A color map is a gradient of colors where each value from 0 to 100 is associated with a color on the gradient. We can choose a color map, and associate a number value with each point to specify its color. We can include the colorbar with the statement plt.colorbar(). More colorbars are available here: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='plasma')

plt.colorbar()

plt.show()
```

We can also individually assign the sizes of markers.

```python
sizes=np.array([50,30,40,90,50,15,30,40,2000,70,150,80,250])
plt.scatter(x, y, s=sizes)
plt.show()
```

We can modify the transparency of the dots with the alpha parameter.

```python
x=np.random.normal(loc=0, scale = 10, size=200)
y=np.random.normal(loc=0, scale = 10, size=200)
plt.scatter(x,y,alpha=0.2, s=100)
plt.show()

colors = np.random.randint(100, size=(200))
sizes = 10 * np.random.randint(100, size=(200))
plt.scatter(x,y,c=colors, s=sizes, alpha=0.2, cmap="plasma")
plt.show()
```

We can create bar graphs with the plt.bar() function.

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([5,7,3,4])
plt.bar(x,y, width=0.95)
plt.show()

plt.barh(x,y, height=0.5, color="LightSeaGreen")
plt.show()
```

We can create a frequency histogram with the hist() function.

```python
x=np.random.normal(loc=100, scale = 20, size=250)
plt.hist(x)
plt.show()
```

We can create a pie chart with the plt.pie() function.

```python
x=np.array([5,7,2,3])
pieLabels=np.array(['A', 'B', 'C', 'D'])
plt.pie(x, labels=pieLabels)
plt.show()

x=np.array([5,7,2,3])
pieLabels=np.array(['A', 'B', 'C', 'D'])
pieExplode = np.array([0, 0, 0.15, 0])
pieColors=np.array(["LightSeaGreen", "hotpink", "coral", "DarkViolet"])
plt.pie(x, startangle=180, explode=pieExplode, shadow=True, colors=pieColors)
plt.legend(title = "Legend:", labels=pieLabels)
plt.show()
```
